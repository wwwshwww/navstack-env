```python
import numpy as np
import gym
import navstack_gym
import matplotlib.pyplot as plt
```

```python
env_id = 'VisibleTreasureHunt-v0'
```

```python
# np.random.seed(42)
env = gym.make(env_id, map_unk_val=50)
env.seed(100)
obs = env.reset()
```

```python
for i in range(5):
    obs, reward, _, _ = env.step(env.action_space.sample())
    if reward > 0:
        print(reward)
```
