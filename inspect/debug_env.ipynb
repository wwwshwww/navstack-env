```python
import numpy as np
import gym
import navstack_gym
import matplotlib.pyplot as plt
```

```python
env_id = 'VisibleKeyHunt-v0'
```

```python
np.random.seed(42)
env = gym.make(env_id)
env.seed(100)
obs = env.reset()
```

```python
for i in range(100):
    obs, reward, _, _ = env.step(env.action_space.sample())
    if reward > 0:
        print(reward)
```

```python
plt.imshow(env.render('rgb_array'))
```

```python
om = np.copy(env.actioner.occupancy_map)
om[om==100] = 1
om[om==-1] = 50
om[om==0] = 100
plt.imshow(om, cmap='gray')
```
